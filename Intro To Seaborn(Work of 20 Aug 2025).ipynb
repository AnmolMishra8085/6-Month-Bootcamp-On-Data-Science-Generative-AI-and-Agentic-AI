# Seaborn
### # Seaborn is a powerful Python data visualization library built on top of Matplotlib, designed for creating attractive and informative statistical graphics.
### # It simplifies the process of visualizing complex datasets and integrates seamlessly with pandas data structures.

### * High-Level Interface: Seaborn provides a user-friendly interface for creating a variety of statistical plots with minimal code.

### * Integration with Pandas: It works well with pandas DataFrames, allowing for easy manipulation and visualization of data.

### * Beautiful Default Styles: The library comes with aesthetically pleasing default styles and color palettes, enhancing the visual appeal of the plots.

### * Statistical Functions: Seaborn includes built-in functions for statistical analysis, making it easier to visualize distributions, relationships, and trends in data.

### * Complex Visualizations: It supports complex visualizations like heatmaps, violin plots, and pair plots, which are useful for exploring data relationships.

## Common Use Cases

### * Exploratory Data Analysis (EDA): Seaborn is widely used for EDA to understand data distributions and relationships before applying machine learning models.

### * Statistical Graphics: It is ideal for creating statistical graphics that help in interpreting data insights effectively.

### * Data Presentation: Seaborn is often used in reports and presentations due to its ability to produce publication-quality visuals.

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize']=6,2
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\DeLL\Downloads\data.csv')
df
```

```python
ves1 = sns.distplot(df['InternetUsers'])
plt.show()
```

```python
ves2 = sns.displot(df['InternetUsers'])
plt.show(ves2)
```

```python
ves3 = sns.displot(df['InternetUsers'],bins = 15)
plt.show(ves3)
```

```python
vis4 = sns.boxplot (data=df,x = 'IncomeGroup', y= 'BirthRate')
plt.show(vis4)
```

```python
# outlier - stats outlier is the data point which is very far from other observation 
# how to detect a outlier -- using graph
# outlier also called as = ANOMALY DETECTION
```

```python
vis7 =sns.lmplot(data = df, x = 'InternetUsers', y= 'BirthRate')
plt.show(vis7)
```

```python
vis5 =sns.lmplot(data = df, x = 'InternetUsers', y= 'BirthRate',fit_reg = False, hue = 'IncomeGroup')
plt.show(vis5)
```

```python
vis5 =sns.lmplot(data = df, x = 'InternetUsers', y= 'BirthRate',fit_reg = True, hue = 'IncomeGroup')
plt.show(vis5)
```
