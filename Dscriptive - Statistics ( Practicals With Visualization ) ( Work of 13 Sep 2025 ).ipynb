# 13.Plot the Histogram to show of Monthly Household Incomeme of peoples

```python
plt.figure(figsize=(6,3))
sns.histplot(df['Mthly_HH_Income'],bins =10, kde=True, color='blue')
plt.title('Distribution of Monthly Household Income', fontsize=16)
plt.xlabel('Monthly Household Income', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

# 14.Plot the Histogram to show of Monthly Household Expense of peoples

```python
plt.figure(figsize=(6,3))
sns.histplot(df['Mthly_HH_Expense'],bins =10, kde=True , color='green')
plt.title('Distribution of Monthly Household Expense', fontsize=16)     
plt.xlabel('Monthly Household Expense', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()  
```

# 15.Plot the Scatter Plot : Income Vs. Expenses

```python
plt.figure(figsize=(6,3))
sns.regplot(x='Mthly_HH_Income', y='Mthly_HH_Expense', data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Monthly Income vs Monthly Expense', fontsize=16)
plt.xlabel('Monthly Household Income', fontsize=12)
plt.ylabel('Monthly Household Expense', fontsize=12)        
plt.show()
```

# 16. Plot the Boxen Plot : No. of Family Members Vs. Monthly Household Expense

```python
plt.figure(figsize=(6,3))
sns.boxenplot(x='No_of_Fly_Members', y='Mthly_HH_Expense', data=df)
plt.title('Monthly Household Expense by Number of Family Members', fontsize=16)
plt.xlabel('Number of Family Members', fontsize=12) 
plt.ylabel('Monthly Household Expense', fontsize=12)
plt.show()
```

# 17.Plot the Pairplot to show the relationship between all the financial variables

```python
sns.pairplot(df[['Mthly_HH_Income', 'Mthly_HH_Expense', 'No_of_Fly_Members','Emi_or_Rent_Amt', 'Annual_HH_Income', 'No_of_Earning_Members']])
plt.suptitle('Pairplot of Financial Variables', y=1.02, fontsize=18)
plt.show()
```

# 18. Plot the Line Plot : Income Vs. Expenses

```python
plt.figure(figsize=(6,3))
plt.plot(df['Mthly_HH_Income'], df['Mthly_HH_Expense'], 'o', linestyle='dashed', color='blue')
plt.title('Monthly Income vs Monthly Expense', fontsize=16)
plt.xlabel('Monthly Household Income', fontsize=12)
plt.ylabel('Monthly Household Expense', fontsize=12)        
plt.show()
```

# 19. Plot the Heatmap to show Correlation between Numerical Variables

```python
numerical_df = df[['Mthly_HH_Income', 'Mthly_HH_Expense', 'No_of_Fly_Members', 'Emi_or_Rent_Amt', 'Annual_HH_Income', 'No_of_Earning_Members']]
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(6,3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Financial Data', fontsize=16)
plt.show()
```
