```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
x1 = np.linspace(0, 10, 50)
plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')
plt.plot(x1, np.tan(x1), '--')
plt.show()
```

```python
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1), '*')
```

```python
plt.plot([0,1,2,3,4,5,6],[1,2.5,3,3.2,2.8,2,1.8],'-*', c = 'r',ms = '20')
plt.xlabel('Time( In Decade )')
plt.ylabel('Helth')
plt.title('Growth And Decline Of Human Helth Condition With Time ')
plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'r-')
ax[1].plot(x1, np.cos(x1), 'b-');
```

### # plt.subplot() : The plt.subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to specify the layout of the subplots and select which subplot to activate for plotting.

### # Parameters of plt.subplot()
### * The function takes three main parameters:

#### *nrows: The number of rows in the subplot grid.
#### *ncols: The number of columns in the subplot grid.
#### *index: The index of the subplot to activate, counting from 1 (top-left) to nrows * ncols (bottom-right).

```python
a1 = plt.subplot(2,2,1)
plt.plot([1, 3, 2, 4], 'b-')
a2 = plt.subplot(2,2,2)
plt.plot([1, 3, 2, 4], 'b-')
a3 = plt.subplot(2,2,3)
plt.plot([0, 5, 2, 4], 'b-')
a4 = plt.subplot(2,2,4)
plt.plot([0, 5, 2, 4], 'b-')
```

```python
x4 = range(1, 5)

plt.plot(x4, [xi*1.5 for xi in x4])

plt.plot(x4, [xi*3 for xi in x4])

plt.plot(x4, [xi/3.0 for xi in x4])

plt.show()
```

### # plot : Plot y versus x as lines and/or markers.

### # marker :  a marker is a symbol used to represent individual data points in plots.
#### * Types of Markers: Matplotlib offers various marker styles, including:

##### '.' : Point
##### 'o' : Circle
##### 's' : Square
##### '^' : Triangle
##### 'D' : Diamond
##### '*' : Star
##### '+' : Plus
##### 'x' : Crossividual data points in plots.


### # Control colours (c) : We can draw different lines or curves in a plot with different colours.

### # legend : Legends are used to describe what each line or curve means in the plot.

### # ms (marker size): Controls the size of the marker.

### # grid :  a grid is a set of horizontal and vertical lines that help to visually align and interpret data points on a plot, making it easier to read values and understand the graph's structure.

### # axis : an axis refers to the reference lines (x-axis and y-axis) on a plot that define the scale and orientation for displaying data, allowing for the measurement of values along horizontal and vertical dimensions.

### # Label: A general term for text descriptions in a plot that provide context or information about the data.

### # x-label: The label for the x-axis, indicating what the horizontal axis represents (e.g., time, categories).

### # y-label: The label for the y-axis, indicating what the vertical axis represents (e.g., values, measurements).

### # title : A title is a text description displayed at the top of a plot that summarizes the main theme or purpose of the graph, providing context and helping viewers understand what the data represents.

### # plt.show() : plt.show() is a function that displays the current figure or plot window, rendering all the visual elements (such as lines, markers, titles, and labels) so that the user can view the generated graph.

```python
plt.plot([0,1,2,3,4,5,6],[1,2.5,3,3.2,2.8,2,1.8],'-*', c = 'r',ms = '20')
plt.axis([-1, 7, 0, 5]) 
plt.legend(['Helth Graph'])
plt.grid(True)
plt.xlabel('Time( In Decade )')
plt.ylabel('Helth')
plt.title('Growth And Decline Of Human Helth Condition With Time ')
plt.show()
```

## Histogram

## Bar Chart

```python
data2 = [5. , 25. , 50. , 20.]

plt.bar(range(len(data2)), data2)

plt.show() 
```

## Horizontal Bar Chart

```python
data2 = [5. , 25. , 50. , 20.]

plt.barh(range(len(data2)), data2)

plt.show() 
```

## Error Bar Chart

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show();
```

## Stacked Bar Chart

```python
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

## Pie Chart

```python
plt.figure(figsize=(7,7))

x10 = [35, 25, 20, 20]

labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']

plt.pie(x10, labels=labels);

plt.show()
```

## Boxplot

```python
data3 = np.random.randn(100)

plt.boxplot(data3)

plt.show();
```

## Area Chart

```python
x12 = range(1, 6)
y12 = [1, 4, 6, 8, 4]

# Area plot
plt.fill_between(x12, y12)
plt.show()
```
