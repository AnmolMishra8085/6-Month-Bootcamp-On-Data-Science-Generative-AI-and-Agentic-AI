```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X =df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor  # <-- IMPORT THE REGRESSOR
from sklearn.metrics import r2_score           # <-- IMPORT NEW METRIC
```

```python
# Create the scaler
scaler = StandardScaler()

# Fit ONLY on the training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data
X_test_scaled = scaler.transform(X_test)
```

```python
# 1. Create the ANN Regressor model
# We set max_iter=1000 to give it more time to train
ann_model = MLPRegressor(random_state=0, max_iter=1000)

# 2. Fit the model on the SCALED training data (This will work now)
ann_model.fit(X_train_scaled, y_train)

# 3. Predict on the SCALED test data
y_pred_ann = ann_model.predict(X_test_scaled)

# 4. Check the R² score (not accuracy)
# R² score of 1.0 is a perfect fit. 0 is a bad fit.
score_ann = r2_score(y_test, y_pred_ann)

print(f"ANN (MLPRegressor) R² Score: {score_ann}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 1. Create the model
lr_model = LinearRegression()

# 2. Fit on the original (unscaled) X_train
lr_model.fit(X_train, y_train)

# 3. Predict on the original (unscaled) X_test
y_pred_lr = lr_model.predict(X_test)

# 4. Get the R² score
score_lr = r2_score(y_test, y_pred_lr)
print(f"Linear Regression R² Score: {score_lr}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# 1. Create the model
rf_model = RandomForestRegressor(random_state=0)

# 2. Fit on the original (unscaled) X_train
rf_model.fit(X_train, y_train)

# 3. Predict on the original (unscaled) X_test
y_pred_rf = rf_model.predict(X_test)

# 4. Get the R² score
score_rf = r2_score(y_test, y_pred_rf)
print(f"Random Forest R² Score: {score_rf}")
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

# 1. Create the SVR model
svr_model = SVR(kernel='rbf') # 'rbf' is a common and powerful kernel to start with

# 2. Fit on the SCALED training data
svr_model.fit(X_train_scaled, y_train)

# 3. Predict on the SCALED test data
y_pred_svr = svr_model.predict(X_test_scaled)

# 4. Get the R² score
score_svr = r2_score(y_test, y_pred_svr)
print(f"SVR R² Score: {score_svr}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

# 1. Create the Gradient Boosting model
gbr_model = GradientBoostingRegressor(random_state=0)

# 2. Fit on the ORIGINAL training data
gbr_model.fit(X_train, y_train)

# 3. Predict on the ORIGINAL test data
y_pred_gbr = gbr_model.predict(X_test)

# 4. Get the R² score
score_gbr = r2_score(y_test, y_pred_gbr)
print(f"Gradient Boosting R² Score: {score_gbr}")
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

# 1. Create the KNN model (we'll start with k=5 neighbors)
knn_model = KNeighborsRegressor(n_neighbors=5)

# 2. Fit on the SCALED training data
knn_model.fit(X_train_scaled, y_train)

# 3. Predict on the SCALED test data
y_pred_knn = knn_model.predict(X_test_scaled)

# 4. Get the R² score
score_knn = r2_score(y_test, y_pred_knn)
print(f"K-Neighbors (KNN) R² Score: {score_knn}")
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# You already have y_test and y_pred_lr from the Linear Regression model

# 1. Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred_lr)
print(f"Mean Squared Error (MSE): {mse}")

# 2. Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
