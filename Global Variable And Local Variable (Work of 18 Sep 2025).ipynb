## 3) global keyword :
### * Used inside a function to modify a global variable.

### * Without global, Python assumes youâ€™re working with a local variable.

### ðŸ”¹ Example:

```python
a = 10  

def something():
    global a 
    b = 15 # 15 is converted to local when user assigned global a 
    print('in function',b)
    print('gloabl variable', a)
    
something()

print('out function',a)
```

```python
x = 10  # Global variable

def update_x():
    global x  # Declare that we are using the global variable x
    x += 10    # Modify the global variable

update_x()
print(x)  # Output: 15
```

## globals() Function

### * Returns a dictionary of all global variables in the program.

### * Can be used to access or modify global variables directly.

### ðŸ”¹ Example:

```python
x = 10  # Global variable

def update_x():
    globals()['x'] += 20  # Access and modify the global variable using the dictionary

update_x()
print(x)  # Output: 15
```

# How to pass list to a function

```python
def count(lst):
    
    even = 0
    odd = 0
    
    for i in lst:
        if i%2 == 0:
            even += 1
        else:
            odd +=1 
    return even,odd

lst = [10, 9, 8, 23, 50, 8, 9, 100 ]
even, odd = count(lst)         

print(even)
print(odd)
```

## Recursion 

### * Definition: A function that calls itself to solve a smaller version of the same problem.

### * Must have a base condition to stop recursion (otherwise infinite loop).

```python
def wish():
    print('hello')
    print('hi')

wish()
```

```python
def wish():
    print('hello')
    print('hi')
    wish()

wish()
```

## Fibonacci Series :

```python
def feb(n):
    a, b = 0, 1
    sequence = []
    for i in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence

x = feb(10)
print("Fibonacci Series up to 10 steps:", x)
```
