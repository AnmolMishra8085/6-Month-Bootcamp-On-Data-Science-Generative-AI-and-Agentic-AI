## Treating Missing Values

```python
clean_dt
```

```python
clean_dt.isnull()
```

```python
clean_dt.isnull().sum()
```

```python
import numpy as np
```

```python
clean_dt['Age']
```

```python
clean_dt['Age'] = clean_dt['Age'].fillna(np.mean(pd.to_numeric(clean_dt['Age'])))
```

```python
clean_dt['Age']
```

```python
clean_dt['Age'] = clean_dt['Age'].astype(int)
```

```python
clean_dt['Age']
```

```python
clean_dt['Salary']
```

```python
clean_dt['Salary'] = clean_dt['Salary'].fillna(np.mean(pd.to_numeric(clean_dt['Salary'])))
```

```python
clean_dt['Salary']
```

```python
clean_dt['Salary'] = clean_dt['Salary'].astype(int)
```

```python
clean_dt['Salary']
```

```python
clean_dt
```

```python
clean_dt['Experience'] = clean_dt['Experience'].fillna(np.mean(pd.to_numeric(clean_dt['Experience'])))
```

```python
clean_dt['Experience']
```

```python
clean_dt['Experience'] = clean_dt['Experience'].astype(int)
```

```python
clean_dt['Experience']
```

```python
clean_dt
```

```python
clean_dt['Department']
```

```python
clean_dt['Department'] = clean_dt['Department'].fillna(clean_dt['Department'].mode()[0])
```

```python
clean_dt['Department']
```

### Type Casting

```python
clean_dt
```

```python
clean_dt['Employee_ID'] = clean_dt['Employee_ID'].astype(int)
clean_dt['Age'] = clean_dt['Age'].astype(int)
clean_dt['Salary'] = clean_dt['Salary'].astype(int)
clean_dt['Experience'] = clean_dt['Experience'].astype(int)
clean_dt['Name'] = clean_dt['Name'].astype('category')
clean_dt['Department'] = clean_dt['Department'].astype('category')
```
